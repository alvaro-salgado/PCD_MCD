```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd
```

```python
# Cargar los audios
audio_perro, sr_perro = librosa.load("/content/dog_barking_99.wav", sr=None)
audio_gato, sr_gato = librosa.load("/content/cat_49.wav", sr=None)
```

```python
# Reproducir los audios
print("Audio de perro:")
ipd.display(ipd.Audio(audio_perro, rate=sr_perro))

print("Audio de gato:")
ipd.display(ipd.Audio(audio_gato, rate=sr_gato))
```

```python
# Función para graficar forma de onda
def plot_waveform(audio, sr, title):
    plt.figure(figsize=(8, 3))
    librosa.display.waveshow(audio, sr=sr, alpha=0.7)
    plt.title(title)
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Amplitud")
    plt.show()
```

```python
# Graficar formas de onda
plot_waveform(audio_perro, sr_perro, "Forma de Onda - Aullido de Perro")
plot_waveform(audio_gato, sr_gato, "Forma de Onda - Maullido de Gato")
```

```python
# Función para graficar espectrograma
def plot_spectrogram(audio, sr, title):
    plt.figure(figsize=(10, 3))
    S = librosa.amplitude_to_db(np.abs(librosa.stft(audio)), ref=np.max)
    librosa.display.specshow(S, sr=sr, x_axis="time", y_axis="log")
    plt.colorbar(format="%+2.0f dB")
    plt.title(title)
    plt.show()
```

```python
# Graficar espectrogramas
plot_spectrogram(audio_perro, sr_perro, "Espectrograma - Aullido de Perro")
plot_spectrogram(audio_gato, sr_gato, "Espectrograma - Maullido de Gato")
```

```python
# Función para graficar la Transformada de Fourier
def plot_fft(audio, sr, title):
    fft = np.fft.fft(audio)
    freqs = np.fft.fftfreq(len(fft), 1 / sr)
    magnitudes = np.abs(fft)

    plt.figure(figsize=(10, 3))
    plt.plot(freqs[:len(freqs)//2], magnitudes[:len(magnitudes)//2])
    plt.title(title)
    plt.xlabel("Frecuencia (Hz)")
    plt.ylabel("Magnitud")
    plt.show()
```

```python
# Graficar FFT
plot_fft(audio_perro, sr_perro, "Espectro de Frecuencia - Aullido de Perro")
plot_fft(audio_gato, sr_gato, "Espectro de Frecuencia - Maullido de Gato")
```
